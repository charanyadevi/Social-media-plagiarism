```python
#Import the Necessary Packages
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
#Load the Data
df = pd.read_csv('/content/drive/MyDrive/Twitter_Data.csv')
```

```python
#Removing the Categorial Data
df.drop(columns=['category'],inplace=True)
```

```python
df
```

```python
#Renaming the Column
df.rename(columns={'clean_text':'text'},inplace=True)
df.head()
```

```python
#converting all the text into lowercase
df['text']= df['text'].str.lower()
df
```

```python
#Handling the missing value
rows_with_null = df[df.isnull().any(axis=1)]
rows_with_null
```

```python
index_to_remove = [148,158694,159443,160560]
df = df.drop(index_to_remove)
```

```python
df = df.reset_index(drop=True)
```

```python
#Removing the punctuation
import string
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

# Apply the function to the text column
df['text'] = df['text'].apply(remove_punctuation)
df
```

```python
#Taking the sample text to check the Plagiarism
text1 = '.'.join(df.iloc[0:50]['text'])
text1
```

```python
# Importing SequenceMatcher from difflib module
from difflib import SequenceMatcher

text2 = input("Enter a text2: ")

# Using the SequenceMatcher
match = SequenceMatcher(None, text1, text2)

# Convert the above output into a ratio and multiply it by 100
result = match.ratio() * 100

# Display the final result
print('Plagiarism occurs:', int(result), "%")
```
